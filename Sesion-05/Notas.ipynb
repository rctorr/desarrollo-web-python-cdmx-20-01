#### EJEMPLO: La definición de una clase y la creación de objetos de esta clase.



Escribe el código en Python para definir la `clase` del objeto definido en el ejemplo anterior, crea un objeto, imprime el objeto e imprime cada una de sus propiedades.

````python
class Producto:
    def __init__(self, id, nombre, cantidad, precio):
        """ Función constructora del objeto Producto """
        self.id = id
        self.nombre = nombre
        self.cantidad = cantidad
        self.precio = precio
        
    def __str__(self):
        """ Regresa la representación en str del objeto """
        return self.nombre

lista_productos = []
prod1 = Producto(1, "Camisa manga larga", 2, 450.00)
print(prod1)
lista_productos.append(prod1)

prod2 = Producto(2, "Playera", 1, 550.00)
lista_productos.append(prod2)
print(prod2)

print(lista_productos)

print(prod1.id)
print(prod1.nombre)
print(prod1.cantidad)
print(prod1.precio)

print(lista_productos[1].precio)
````

### Los decoradores en las clases de Python
---
#### EJEMPLO:

En la clase del ejemplo para el carrito de la tienda en línea también se desea conocer el subtotal de cada producto, modificar la definición de la clase para incluirlo.

````python
class Producto:
    def __init__(self, id, nombre, cantidad, precio):
        """ Función constructora del objeto Producto """
        self.id = id
        self.nombre = nombre
        self.cantidad = cantidad
        self.precio = precio
        
    def __str__(self):
        """ Regresa la representación en str del objeto """
        return self.nombre
    
    @property
    def subtotal(self):
        """ Calcula el subtotal del producto """
        return self.cantidad * self.precio

    
prod1 = Producto(1, "Camisa manga larga", 2, 450.00)
print(prod1.id, prod1, prod1.cantidad,
      prod1.precio, prod1.subtotal )
````
